```python
def partition(nums, start, end):
    '''
    1. 
    2. 
    3. i=j
    4. nums
    5. ijijQuickSort
    
    '''
    i, j = start, end
    tmp = nums[i] # nums[i]
    while i<j:
        while j>i and nums[j]>=tmp: # 
            j -= 1                     ## 
        nums[i] = nums[j]              ## i
        while i<j and nums[i]<=tmp: # 
            i += 1
        nums[j] = nums[i]
    nums[i] = tmp
    return i

def QuickSort(nums, start, end):
    """
    \
    
    nums
    start
    end
    """
    # 
    if start>=end:
        return nums
    else:
        i = partition(nums, start, end)
        QuickSort(nums, start, i-1)
        QuickSort(nums, i+1, end)
    return nums
```
